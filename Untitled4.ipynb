## Implement data storage and security

### Subtask:
Set up the on-device encrypted data silos for secure storage of case data, ensuring strict separation between cases.

**Reasoning**:
Outline the requirements for on-device data storage, describe the encryption approach, explain how data silos will be implemented, detail data access methods, and discuss backup/export, then finish the subtask.

```python
# 1. Outline the requirements for on-device data storage
storage_requirements = """
**On-Device Data Storage Requirements:**

-   **Security:** This is paramount. All sensitive case data must be stored securely on the user's device.
-   **Encryption:** All case data at rest must be encrypted to protect it from unauthorized access if the device is lost or compromised.
-   **Data Silos:** Each case must have its own completely separate and isolated storage area (silo) to prevent cross-contamination of information between cases. The application must only be able to access data for the currently open case.
-   **Data Types:** The storage system must accommodate both structured data (e.g., timeline events, entity profiles, deadlines, citations) and unstructured data (e.g., uploaded documents, images, audio files).
-   **Reliability:** Data storage must be robust and prevent data loss.
-   **Performance:** Accessing and writing data should be reasonably fast to ensure a smooth user experience.
-   **Scalability (within device limits):** The system should be able to handle a significant amount of data per case, up to the storage capacity of the user's device.
"""
print(storage_requirements)

# 2. Describe the chosen approach for implementing encryption
encryption_approach = """
**Encryption Implementation Approach:**

To ensure the security of sensitive case data stored on the user's device, strong encryption at rest is essential.

**Approach:**

1.  **Generate a Case-Specific Encryption Key:** When a new case is created, a unique, strong encryption key will be generated specifically for that case. This key will be used to encrypt all data within that case's silo.
2.  **Key Management:** The case-specific encryption key itself must be protected.
    *   It should *not* be stored in plain text on the device.
    *   A common approach for mobile/desktop applications is to derive or encrypt this case key using a key derived from the user's authentication credentials (e.g., a password, PIN, or biometric authentication). This means the user must authenticate to unlock access to their case data.
    *   Platform-specific secure storage (e.g., Android Keystore, iOS Keychain, Windows Credential Manager) should be used to store the encrypted case key or the key used to derive it.
3.  **Data Encryption:**
    *   All structured data (e.g., stored in a local database like SQLite) should be encrypted before being written to disk. Database encryption features (like SQLCipher for SQLite) are highly recommended.
    *   All unstructured files (documents, images, etc.) should be encrypted using the case-specific key before being saved to the file system. Standard symmetric encryption algorithms (e.g., AES-256) should be used.
4.  **Decryption on Access:** When the application needs to access data for the currently open case, the user must authenticate to unlock the case key. This key is then used to decrypt the data read from the database or files before it is processed by the Application Logic or AI Processing layers.
5.  **No Central Key Storage:** Crucially, no encryption keys should be stored on any central server. Keys are generated and managed *only* on the user's device, linked to their local authentication.
"""
print(encryption_approach)

# 3. Explain how data silos will be implemented
data_silos_implementation = """
**Data Silos Implementation:**

Implementing strict data silos ensures that data from one case cannot be accidentally or maliciously accessed when another case is open.

**Approach:**

1.  **Case-Specific Directories/Containers:** On the device's file system, a unique, dedicated directory or container will be created for each case. The name of this directory should ideally be derived from the unique, AI-generated Internal Case ID to avoid naming conflicts and provide a clear mapping.
2.  **Separate Storage Instances:**
    *   **Structured Data:** Each case will have its own separate instance of the local database (e.g., a unique SQLite database file per case). This database file will store all the structured data for that specific case (timeline, entities, citations, etc.) and will be encrypted using the case's unique key.
    *   **Unstructured Data:** All uploaded files (documents, images, etc.) should be encrypted using the case-specific key before being saved to the file system. Standard symmetric encryption algorithms (e.g., AES-256) should be used.
3.  **Application Logic Enforcement:** The Application Logic Layer is responsible for enforcing the silo separation.
    *   When a user opens a case, the Application Logic loads *only* the encryption key and the storage instance (database file and directory path) associated with that specific case ID.
    *   All subsequent data read/write operations initiated by the Application Logic or AI Processing components must be strictly confined to that case's directory/database instance. The application should not have access paths or loaded instances for other cases simultaneously.
    *   Switching cases involves closing the current case's storage instances and key, and then opening the selected new case's instances and key after user authentication if required.
4.  **No Cross-Linking:** Prevent any direct links or references between data entities stored in different case silos at the storage level. Relationships (e.g., linking an entity to a fact) should only exist *within* a single case silo.
"""
print(data_silos_implementation)

# 4. Detail the methods for accessing and managing data within these encrypted silos
data_access_management = """
**Data Access and Management within Encrypted Silos:**

Accessing data requires decryption, and management involves ensuring data integrity and security.

**Access Methods:**

1.  **Authentication Required:** Access to any case's data requires user authentication to unlock the case-specific encryption key.
2.  **Application Logic as Gateway:** All data access (read and write) should be mediated by the Application Logic Layer. AI Processing components should request data from or send data to the Data Storage Layer *through* the Application Logic, which handles the encryption/decryption and ensures operations are confined to the correct case silo.
3.  **Database Access:** When the database for the current case is accessed, the Application Logic uses the unlocked case key to decrypt the database file (e.g., using the SQLCipher library's API). Standard database queries are then performed on the decrypted data in memory before re-encrypting on write.
4.  **File Access:** When a specific file is needed (e.g., to display a document preview or for AI analysis), the Application Logic:
    *   Locates the encrypted file within the current case's directory.
    *   Reads the encrypted file content.
    *   Uses the unlocked case key to decrypt the content.
    *   Provides the decrypted content to the requesting component (e.g., UI for display, AI Processing for analysis). Decrypted content should be kept in memory only as long as needed and not written to temporary unencrypted files.

**Management Considerations:**

1.  **Data Integrity:** Implement mechanisms (like database transactions) to ensure data consistency during write operations.
2.  **Error Handling:** Gracefully handle potential errors during encryption/decryption or file access.
3.  **Updates:** Plan for how the schema of the structured data (database) might evolve and how updates will be managed while preserving encrypted data.
4.  **Secure Deletion:** When a case is deleted, ensure that the case's encryption key is securely erased and all encrypted files and the database file within that case's silo are securely deleted (e.g., by overwriting the data before deletion, depending on platform capabilities).
"""
print(data_access_management)

# 5. Discuss the backup and export functionality
backup_export_discussion = """
**Backup and Export Functionality in the Context of Encrypted On-Device Storage:**

The 'Export Full Case File' feature (Module 6 in the refined list) is crucial for allowing users to back up their data and maintain control.

**Approach for Backup/Export:**

1.  **User-Initiated Process:** Export is triggered explicitly by the user via the UI.
2.  **Data Collection:** The Application Logic collects all data for the current case from its silo in the Data Storage Layer (structured data from the database, all unstructured files). This data is accessed and decrypted in memory.
3.  **Packaging:** The collected decrypted data is organized into a structured format (e.g., a set of directories and files within a temporary location).
4.  **Re-Encryption for Export:** The entire package of case data should be re-encrypted for the export file.
    *   **Method 1 (Recommended for Backup):** Encrypt the ZIP file (or other archive format) using a *new*, user-defined password. This allows the user to store the backup file anywhere and decrypt it later using the password, independent of the original device or authentication method. Standard, strong encryption for ZIP/archives should be used.
    *   **Method 2 (Less Recommended for Portability):** Encrypt the ZIP file using the *case-specific* key. This requires access to the original application setup and potentially the original authentication method to decrypt the backup, making it less portable. Method 1 is generally preferred for user-controlled backups.
5.  **File Format:** Package the encrypted data into a single archive file (e.g., a password-protected ZIP or a custom encrypted container format).
6.  **User Destination:** Allow the user to choose where to save the exported, encrypted file (e.g., local storage, cloud drive via OS picker).
7.  **Import Functionality (Reverse Process):** A corresponding import function is needed to restore a case from an exported, encrypted file. This would involve:
    *   User selecting the encrypted backup file.
    *   Prompting the user for the decryption password (if Method 1 was used).
    *   Decrypting the archive.
    *   Creating a *new* case silo on the device with a new unique ID and key.
    *   Importing the data from the decrypted archive into the new case silo, encrypting it with the new case key.
    *   Handling potential conflicts if a case with the same original ID already exists.

This process ensures that users can securely back up and export their sensitive legal data while maintaining control through encryption.
"""
print(backup_export_discussion)
```

```python
# 1. AI Counsel Persona Selector Design and Implementation
persona_selector_design = """
**AI Counsel Persona Selector**

**User Interface (UI) Design:**
-   **Access:** Likely a menu item in settings or potentially accessible from the persistent AI Bot interface.
-   **Layout:** A dedicated screen or modal dialog.
-   **Components:**
    -   Heading: "Choose Your AI Counsel Persona"
    -   Description: Brief explanation of what personas are and how they affect tone, not accuracy.
    -   Persona List/Cards: A visual list where each persona is represented (e.g., "The Strategist," "The Guide"). Each entry could include:
        -   Persona Name
        -   Brief description of the persona's tone/style (e.g., "Direct and analytical," "Supportive and empathetic").
        -   (Optional) A small icon or image representing the persona.
        -   A "Select" button or tap area for each persona.
        -   (Optional) A "Preview" button that shows example responses from the AI in that persona's style.
    -   Confirmation/Save Button: A button to confirm the selected persona if not applied immediately on selection.

**User Flow:**
1.  User navigates to the Persona Selector interface.
2.  User browses the available personas and reads their descriptions.
3.  (Optional) User previews responses in different persona styles.
4.  User selects their desired persona.
5.  The application confirms the selection, and the AI's communication style is updated for future interactions.

**Technical Implementation Outline:**
-   **Persona Data:** Store predefined data for each persona (name, description, parameters influencing tone/style).
-   **AI Model Integration:** The underlying AI model(s) must support controllable output parameters (e.g., temperature, top-p, specific prompt instructions) that can modulate the tone, verbosity, and style of the generated text *without* changing the factual or legal core of the response.
-   **Application Logic Role:** The Application Logic Layer will:
    -   Store the user's selected persona preference (in Data Storage, linked to user profile/settings).
    -   Retrieve the active persona when initiating a request to the AI Processing Layer (specifically for text generation, like bot responses or drafting suggestions).
    -   Pass the relevant persona parameters (derived from the stored persona data) as instructions or parameters to the AI model call.
-   **AI Processing Role:** The AI Processing components (especially the AI Counsel Bot and potentially drafting features) will receive the persona parameters and adjust their output generation accordingly.
-   **Ethical Constraint Enforcement:** Crucially, the AI's core programming and 'Red Line' disclaimers must override any persona instruction if the persona's style would lead to providing legal advice or inaccurate information. The persona only affects *how* legally accurate information is communicated, not *what* legal information is communicated.
"""
print(persona_selector_design)

# 2. Focus & Wellness Mode Design and Implementation
wellness_mode_design = """
**Focus & Wellness Mode**

**User Interface (UI) Design:**
-   **Access:** A discrete, always-available toggle button or icon, likely in the Top Bar or a persistent footer/sidebar element.
-   **Simplified UI Toggle:** Clicking the toggle/icon switches the main application interface to a simplified mode.
-   **Simplified UI (Example):**
    -   Reduced visual clutter (e.g., hiding non-essential icons, borders, or complex layouts).
    -   Potentially muted color scheme.
    -   Focus on core task at hand (e.g., if on the timeline, only show the chronological list view, hide the map/matrix).
    -   Prominent display of the AI Counsel Bot for easy access to supportive coaching.
-   **Access to Supportive Content:** Within the simplified mode or via a secondary click on the wellness icon:
    -   A dedicated section or pop-up.
    -   Components:
        -   Heading: "Focus & Wellness Resources"
        -   Supportive AI Coaching Access: A clear button or link to initiate a conversation with the AI Bot in a specific supportive mode.
        -   Curated Content List: Links to external, non-legal resources (e.g., mental health hotlines, stress management guides, mindfulness exercises). These would open in the device's default browser.

**User Flow:**
1.  User feels overwhelmed and clicks the Focus & Wellness Mode toggle.
2.  The application UI immediately switches to a simplified visual state.
3.  User can then access supportive AI coaching via the bot or browse external resources.
4.  User clicks the toggle again to return to the standard UI.

**Technical Implementation Outline:**
-   **UI State Management:** The Application Logic Layer manages the UI state (standard vs. simplified).
-   **CSS/Styling:** Implement two distinct sets of styling (CSS rules, themes) for the UI elements. Toggling the mode applies the simplified styling rules application-wide.
-   **Conditional Rendering/Layout:** Application Logic conditionally hides or rearranges UI components based on the active mode (e.g., hide complex visualization panels in the Factual Record module when in Focus Mode).
-   **AI Bot Mode:** When the user accesses AI coaching from this mode, the Application Logic can pass a specific context or parameter to the AI Bot component, instructing it to adopt a supportive tone and focus the conversation on stress management or well-being, while still adhering to its core principles (e.g., not giving medical advice).
-   **Resource Management:** Store a curated list of external resource links within the application's assets or a dedicated data structure. Ensure these links are safe and open externally.
"""
print(wellness_mode_design)

# 3. Export Full Case File Design and Implementation
export_case_design = """
**Export Full Case File**

**User Interface (UI) Design:**
-   **Access:** A menu item, likely in a main "File" or "Case" menu, or within the 'Post-Filing Management' or 'Multi-Case Management Dashboard'.
-   **Layout:** A dedicated screen or a multi-step modal dialog.
-   **Components:**
    -   Heading: "Export Case File"
    -   Description: Explain that this will create an encrypted backup of the entire case.
    -   Case Selection (if accessed from dashboard): Allow user to select which case to export.
    -   Export Options:
        -   File Format: (Likely pre-selected as encrypted ZIP or custom format). Display this information.
        -   Encryption Method: Explain that it will be password-protected.
    -   Password Input:
        -   Label: "Set Export Password"
        -   Input field for password (with toggle to show/hide password).
        -   Label: "Confirm Password"
        -   Input field to re-enter password for confirmation.
        -   Password Strength Indicator (Optional but Recommended).
        -   Warning: Emphasize that this password is required to open the export file and cannot be recovered by the app if lost.
    -   Destination Selection:
        -   Button: "Choose Save Location" - This should trigger the device's native file picker to select a directory.
        -   Display Area: Show the selected save path.
    -   Action Button: "Export Case"

**User Flow (Step-by-Step):**
1.  User initiates the export process.
2.  (If applicable) User selects the case to export.
3.  User is prompted to set and confirm a password for the export file.
4.  User chooses the destination folder using the native file picker.
5.  User clicks "Export Case."
6.  A progress indicator is shown while the export is in progress.
7.  A confirmation message is displayed upon successful export.

**Technical Implementation Outline:**
-   **Data Collection:** The Application Logic Layer initiates the process. It accesses the Data Storage Layer to retrieve *all* data associated with the selected case's silo:
    -   Retrieve all structured data from the case's database instance.
    -   Retrieve all unstructured files from the case's dedicated directory.
-   **Decryption (In-Memory):** The retrieved data is decrypted using the case's active encryption key *in memory*. Decrypted data should not be written to temporary files on disk.
-   **Packaging:** Organize the decrypted data into a structured format in a temporary memory buffer or a secure, temporary encrypted file on disk (if memory is insufficient). This structure should mirror the case's organization (e.g., folders for documents, a file for timeline data).
-   **Password-Based Encryption:** Use a standard, strong encryption library (e.g., using AES-256) to encrypt the packaged data. The encryption key for this step is derived from the password the user set during the export process. A secure Key Derivation Function (KDF) like scrypt or Argon2 should be used to derive a strong encryption key from the user's password.
-   **Archiving:** Package the encrypted data into a standard archive format like a ZIP file that supports encryption, or a custom container format.
-   **Writing to File:** Write the final encrypted archive file to the location selected by the user using the device's file system APIs.
-   **Error Handling:** Implement robust error handling for issues like insufficient storage space, encryption failures, or file writing errors.
-   **Progress Reporting:** Provide feedback to the UI on the export progress.
-   **Secure Cleanup:** Ensure that any temporary decrypted data or temporary files created during the process are securely erased from memory and disk after the export is complete or if an error occurs.
"""
print(export_case_design)

# 4. Redaction & Sharing Tool Design and Implementation
redaction_sharing_design = """
**Redaction & Sharing Tool**

**User Interface (UI) Design:**
-   **Access:** Accessible from the 'Evidence Locker & Persons & Entities' module when viewing or selecting a document or image file.
-   **Layout:** A dedicated viewing/editing interface for the selected file.
-   **Components:**
    -   Document/Image Viewer: Displays the content of the selected file.
    -   Redaction Tools:
        -   "Redact Area" Tool: Allows the user to draw rectangles over sensitive text or images.
        -   "Redact Text" Tool (Optional but Recommended): Allows the user to select specific text to redact (requires text layer from OCR).
        -   Redaction Mark Style Selector (Optional): Black bar, colored box, or "Redacted" text.
    -   Undo/Redo Buttons.
    -   "Apply Redactions" Button: Processes the file and creates a redacted copy.
    -   "Share Redacted File" Button: Appears after redaction is applied, triggering the device's native sharing options.
    -   (Optional) "Save Redacted Copy" Button: Allows saving the redacted version as a new file within the case silo.
    -   Original vs. Redacted View Toggle (Optional): To switch between the original and the redacted preview.

**User Flow:**
1.  User navigates to the 'Evidence Locker' and selects a document or image.
2.  User selects the "Redact & Share" option.
3.  The file opens in the redaction interface.
4.  User uses the redaction tools to mark areas or select text to be redacted.
5.  User reviews the applied redactions in the preview.
6.  User clicks "Apply Redactions."
7.  The application processes the file and creates a *new* redacted version.
8.  User is presented with options to "Share Redacted File" or (optionally) "Save Redacted Copy."
9.  If sharing, the native share sheet appears. If saving, the redacted file is saved.

**Technical Implementation Outline:**
-   **File Loading and Rendering:** Load the selected file's content. For PDFs or images, render them into a format suitable for display and interaction (e.g., image layers). If a PDF has a text layer (from original or OCR), make the text selectable for the "Redact Text" tool.
-   **Redaction Markup:** The UI layer captures the user's redaction marks (e.g., coordinates and dimensions of rectangles, or selected text ranges). This is initially just markup, not altering the original file.
-   **Applying Redactions (Creating New File):** When the user clicks "Apply Redactions":
    -   Access the *original*, decrypted file content from Data Storage.
    -   Create a *new* file based on the original.
    -   For each redaction mark:
        -   **Image Redaction:** Draw a solid block (e.g., black rectangle) over the specified area on the image representation of the page/file.
        -   **Text Redaction (for PDFs with text layer):** Modify the PDF structure to remove or obscure the selected text from both the visual layer and the underlying text layer to prevent copy-pasting redacted text. Simply drawing a box over text in an image layer is insufficient for secure text redaction in PDFs.
    -   Save the newly created, redacted file to a temporary, secure location.
-   **Original File Integrity:** The *original* file in the case silo must remain unchanged. Redaction *always* creates a new file.
-   **Sharing:** Use the device's native sharing APIs to share the temporary, redacted file. The application should not handle the actual transfer, relying on the OS.
-   **Saving Redacted Copy:** If the user chooses to save, encrypt the temporary redacted file using the case's encryption key and save it as a new file within the case's silo in Data Storage, clearly indicating it's a redacted version.
-   **Secure Cleanup:** Delete the temporary redacted file securely after sharing or saving is complete.
"""
print(redaction_sharing_design)

# 5. Jurisdiction & Court Selector Design and Implementation
jurisdiction_selector_design = """
**Jurisdiction & Court Selector**

**User Interface (UI) Design:**
-   **Access:** Likely part of the 'Case Genesis & Onboarding' process or accessible in Case Settings.
-   **Layout:** A dedicated screen or a step within a case creation wizard.
-   **Components:**
    -   Heading: "Select Jurisdiction and Court"
    -   Description: Explain the importance of selecting the correct court as it affects procedures and rules.
    -   Jurisdiction Selection:
        -   Option Group/Radio Buttons: "State Court (Texas)" vs. "Federal Court."
    -   State Court Specifics (if "State Court" selected):
        -   Dropdown/Searchable List: "Select County" (to identify the specific county court or district court within that county).
        -   (Optional) Further refinement if needed (e.g., District Court vs. County Court vs. Justice Court - depending on case type).
    -   Federal Court Specifics (if "Federal Court" selected):
        -   Dropdown/Searchable List: "Select District" (e.g., Northern District of Texas, Southern District of Texas).
        -   (Optional) Further refinement if needed (e.g., Division).
    -   Educational Info Link: A link or button to access information from the Legal Intelligence module (Knowledge Base) explaining the strategic pros and cons of State vs. Federal court for their case type (as identified during Genesis).
    -   Action Button: "Confirm Selection" or "Continue"

**User Flow:**
1.  User reaches the Jurisdiction & Court Selection step during case creation or edits settings.
2.  User selects either "State Court (Texas)" or "Federal Court."
3.  Based on the jurisdiction, relevant fields appear (County for State, District for Federal).
4.  User selects the specific county or federal district.
5.  (Optional) User reviews educational information on venue selection.
6.  User confirms their selection.
7.  The application saves the selected jurisdiction and court.

**Technical Implementation Outline:**
-   **Jurisdiction Data:** Store structured data about Texas counties, Federal districts in Texas, and their associated court types and basic geographical information.
-   **Application Logic Role:**
    -   Present the correct UI options based on the initial jurisdiction choice (State/Federal).
    -   Capture and validate the user's selection (County/District).
    -   Store the selected jurisdiction and court information in the Data Storage Layer, linked to the specific case.
    -   This stored jurisdiction information is critical and will be accessed by other modules.
-   **Impact on Knowledge Base Access (AI Processing):**
    -   When the AI Processing Layer (e.g., Interview Engine, AI Bot, Litigation & Drafting) needs legal information, it will *always* receive the currently active case's selected jurisdiction and court from the Application Logic.
    -   Queries to the Knowledge Base (especially Tier 2: Specialized Knowledge Base) will be filtered or prioritized based on this jurisdiction. For example, when explaining a legal standard, the AI will prioritize citing Texas statutes and cases if a Texas State court is selected, or Federal law if a Federal court is selected.
    -   The "Texas-First" protocol in the Live Research Assistant (Tier 3) needs to be adaptable or work in conjunction with the selected jurisdiction (e.g., prioritize searches within the selected Federal district's court site if Federal is chosen).
-   **Generating Procedural Checklists:** The Application Logic, using the stored jurisdiction and court information, will query the Specialized Knowledge Base (Tier 2) to retrieve the relevant 'Court-Specific Procedural Checklists' for that specific court. These checklists will be displayed and managed in the 'Litigation & Drafting' or 'Post-Filing Management' modules.
-   **Educational Content:** The Application Logic retrieves the relevant educational content comparing State vs. Federal venue (from the Knowledge Base) and presents it to the user when the educational link is clicked.
"""
print(jurisdiction_selector_design)
```
